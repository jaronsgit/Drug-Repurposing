```python
import py3Dmol
import time
import os

def render_pdb(pdb_id):
    print("Generating target-ligand complex from AlphaFold...")
    time.sleep(2)
    
    pdb_file = f'{pdb_id}.pdb'
    if not os.path.exists(pdb_file):
        print(f"Error: File {pdb_file} not found.")
        return
    
    viewer = py3Dmol.view(width=800, height=600)
    
    with open(pdb_file, 'r') as pdb:
        pdb_content = pdb.read()
    
    viewer.addModel(pdb_content, 'pdb')
    
    viewer.setStyle({'chain': ['A', 'C', 'D', 'P', 'T']}, 
                    {'cartoon': {'color': 'lightgrey', 'opacity': 0.5}})
    
    viewer.setStyle({'hetflag': True}, {'stick': {'colorscheme': 'orangeCarbon'}})
    
    viewer.zoomTo({'hetflag': True}) 
    viewer.zoom(0.5)
    
    return viewer.show()

render_pdb('7l1f')
```
